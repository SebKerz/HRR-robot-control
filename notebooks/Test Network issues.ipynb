### Define helper function for moving
This function will print "we are pretty close" once the position error is < 0.001m and will print "overshoot detected" if the position error is > 0.001m after being pretty close before.

```python
def move_to_test(goal_pose, v_max):
    cobot.move_to_pose(goal_pose, v_max=v_max)
    never_printed = True
    for t in range(int(100 * cobot.hz)):
        if cobot.state is None:
            break
        elif cobot.state == "error":
            break
        if never_printed and np.linalg.norm(goal_pose.t - cobot.sns_pos) < 0.001:
            print("we are pretty close")
            never_printed = False
        if (not never_printed) and np.linalg.norm(goal_pose.t - cobot.sns_pos) > 0.001:
            print("overshoot detected!")
            overshoot_counter+=1
        cobot.update()
```

```python
v_maximale = 0.1
```

```python
overshoot_counter=0
```

### Define two poses for testing

```python
goal_pose1 = cobot.T_B_E_robot.copy()
goal_pose2 = sm.SE3([0,0,-0.3]) @ cobot.T_B_E_robot
```

```python
move_to_test(goal_pose2,v_maximale)
```

```python
move_to_test(goal_pose1,v_maximale)
```

```python
cobot.close_tool_changer()
```

### Move routine to hopefully see errors

```python
for i in range(100):
    move_to_test(goal_pose2,v_maximale)
    move_to_test(goal_pose1,v_maximale)
```
